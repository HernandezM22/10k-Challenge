```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.offline as py
import plotly.graph_objs as go
import plotly.tools as tls
import seaborn as sns
```

```python
df = pd.read_csv("data.csv")
df
```

```python
def filter_by_dates(begin,end,data):
    return data[data[data['dates']==begin].index[0]:data[data['dates']==end].index[0]]


def generate_sphere_viz(data, startdate, enddate, mode = "inline"):

    df = filter_by_dates(startdate, enddate, data)

    countries = df.columns[1:]
    mean_temp = []
    for country in countries:
        mean_temp.append(df[country].mean())

    data = [ dict(
            type = 'choropleth',
            locations = countries,
            z = mean_temp[0:284],
            locationmode = 'country names',
            text = countries,
            marker = dict(
                line = dict(color = 'rgb(0,0,0)', width = 1)),
                colorbar = dict(autotick = True, tickprefix = '', 
                title = '# Average\nTemperature,\n°C')
                )
        ]

    layout = dict(
        title = 'Average temprature change in countries',
        geo = dict(
            showframe = False,
            showocean = True,
            oceancolor = 'rgb(0,255,255)',
            projection = dict(
            type = 'orthographic',
                rotation = dict(
                        lon = 60,
                        lat = 10),
            ),
            lonaxis =  dict(
                    showgrid = True,
                    gridcolor = 'rgb(102, 102, 102)'
                ),
            lataxis = dict(
                    showgrid = True,
                    gridcolor = 'rgb(102, 102, 102)'
                    )
                ),
            )

    fig = dict(data=data, layout=layout)

    if mode == "inline":
        py.iplot(fig, validate=False, filename='spherical_map')
    elif mode =="export":
        py.plot(fig, validate=False, filename='spherical_map')
    else:
        print("Invalid mode")

#------------------------------------------------------------------------------------#
        

def generate_worldmap_viz(data, startdate, enddate, mode = "inline"):

    df = filter_by_dates(startdate, enddate, data)

    countries = df.columns[1:]
    mean_temp = []
    for country in countries:
        mean_temp.append(df[country].mean())

    data = [ dict(
            type = 'choropleth',
            locations = countries,
            z = mean_temp[0:284],
            locationmode = 'country names',
            text = countries,
            marker = dict(
                line = dict(color = 'rgb(0,0,0)', width = 1)),
                colorbar = dict(autotick = True, tickprefix = '',
                title = '# Average\nTemperature,\n°C'),
            #The following line is also needed to create Stream
            #stream = stream_id
                )
        ]
    
    layout = dict(
        title = 'Average land temperature in countries',
        geo = dict(
            showframe = False,
            showocean = True,
            oceancolor = 'rgb(0,255,255)',
            type = 'equirectangular'
        ),
    )
    
    fig = dict(data=data, layout=layout)
    if mode == "inline":
        py.iplot(fig, validate=False, filename='projection_map')
    elif mode =="export":
        py.plot(fig, validate=False, filename='projection_map')
    else:
        print("Invalid mode")
```

```python
generate_sphere_viz(df,'1961-01-31', '2018-12-31', mode = "inline")
```

```python
generate_worldmap_viz(df,'1961-01-31', '2018-12-31', mode="inline")
```
