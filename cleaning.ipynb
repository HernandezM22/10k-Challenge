# Importacion de librerias

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("Environment_Temperature_change_E_All_Data_NOFLAG.csv")
data
```

```python
data = data[data["Element"] == "Temperature change"]
data
```

```python
data = data.drop(columns=["Months Code", "Element Code", "Unit", "Element"])
data
```

```python
data = data[(data["Months"] != "Dec–Jan–Feb") & (data["Months"] != "Mar–Apr–May") & (data["Months"] != "Jun–Jul–Aug") & (data["Months"] != "Sep–Oct–Nov") & (data["Months"] != "Meteorological year")]
data
```

```python
data_test = data.drop(columns=["Area Code", "Months"])
data_test
```

```python
data["Months"].unique()
```

```python
regions = data["Area"].unique()
regions
```

```python
df = pd.DataFrame()
df["dates"] = pd.date_range(start='1/1/1961', end='1/1/2020', freq="M")
df
```

```python
for name in regions:
    df[name] = np.nan

df
```

```python
from datetime import datetime
df['dates'] = pd.to_datetime(df['dates'])
df['Year'] = df['dates'].dt.year
df['Month'] = df['dates'].dt.month 
df
```

```python
frames = []
for name in regions:
    a = data_test[data_test["Area"] == name]
    vect = []
    for column in a.columns:
        if column != "Area":
            vect.extend(a[column])
    df[name] = vect

df
```
